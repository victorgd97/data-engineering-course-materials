```python
#Challenge 1
with open('earthquakes.csv', newline = '') as csvfile:
    reader = csv.reader(csvfile)
    headers = reader.__next__()
    for row in reader:
        magnitude = float(row[2])
        location = row[9]
        time = row[12]
        if magnitude > 6:
            print(magnitude, location, time)
```

```python
#Challenge 2
with open('earthquakes.csv', newline = '') as csvfile:
    reader = csv.reader(csvfile)
    headers = reader.__next__()
    counter = 0
    for row in reader:
        magnitude = float(row[2])
        location = row[9]
        time = row[12]
        if "2016-08-04" in time:
            print(magnitude, location, time)
            counter = counter + 1
    print("There are ", counter, "earthquakes that happened on the date 2016-08-04.")
```
